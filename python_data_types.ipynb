<h1 style="color: gold;">Numbers</h1>

```python
x = 1     # int
y = 2,5   # float
z = 1+2j  # complex

print(type(x))
print(type(y))
print(type(z))

# check data type
a = isinstance(z, complex)
print(a)

# conversion
print(float(22))
print(int(3.14))
print(int("5"))
b = str(x)
print(type(b))
```

<h1 style="color: green;">Boolean</h1>

```python
x = True
y = False

print(x and y)
print(x or y)
```

<h1 style="color: aqua;">List</h1>

```python
x = [1, 2.7, "python", True, 3, "check", 4, 5]
print(x[0])        # first index
print(x[-1])       # last index
print(x[1:5])      # from index 1 to before 5
print(x[:5])       # from first index to before 5
print(x[-4:])      # from -4 index to last
print(x[0:7:2])    # list from index 0 to 7 with step 2

x[2] = "notebook"  # change list element
x.append("plus")   # add list element
del x[3]           # delete list element
print(x)
```

```python
# string slicing
x = "Hello World!"
print(x[0])
print(x[6:])
x = "Hello Bei!"
print(x[6:])

# convert list to set
print(set([1,2,3]))

# convert string to list
print(list('hello'))
```

<h1 style="color: magenta;">Tuple</h1>

```python
x = (5, 'program', 1+3j)
print(x[1])
print(x[0:3])
```

<h1 style="color: Crimson;">Set</h1>

```python
x = {1,2,2,3,3,3}
print(x)
```

```python
# convert set to tuple
print(tuple({5,6,7}))
```

<h1 style="color: Darksalmon;">Dictionary</h1>

```python
x = {1:'value','key':2}
print(type(x))
print("x[1] = ", x[1])
print("x['key'] = ", x['key'])
```

```python
# convert list from tuple to dictionary
print(dict([['key',1],[2,'value']]))
```
